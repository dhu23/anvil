```python
from mbte import simulation, gbm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
help(gbm.GBMParameters)
```

```python
gbm_params = gbm.GBMParameters(0.001, 0.05) 
rng = np.random.default_rng(9)
```

```python
log_returns = gbm.log_returns(gbm_params, 1000, 1, rng)
```

```python
sim_ret = simulation.simulate_prices(1.0, log_returns) 
```

```python
plt.plot(sim_ret.log_returns)
```

```python
plt.plot(sim_ret.prices)
```
